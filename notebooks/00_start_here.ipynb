## What is this, in one sentence?

**Strands Cosmos gives an AI agent two superpowers:** *eyes that understand the physical
world* 👀, and *hands that create video, sound, and robot actions* ✋ — all running on your
own computer.

## Think of it like a brain with two halves

Imagine a really smart student. One half of their brain **watches and understands**
("I see two workers walking past a yellow loader"). The other half **imagines and creates**
("Here's a brand-new video of a robot doing that"). Cosmos has both halves.

## The two model families (pick by your hardware)

| Family | Providers | Think of it as… | Runs on |
|--------|-----------|-----------------|---------|
| 🌟 **Cosmos 3** (newest) | `Cosmos3ReasonerModel`, `Cosmos3GeneratorModel` | The full brain — understands *and* creates | A big GPU (~46 GB) |
| 🪶 **Cosmos-Reason2** (lightweight VLM) | `CosmosVisionModel`, `CosmosModel` | Just the *understanding* half, small & fast | Edge devices (Jetson Thor/Orin) |

!!! tip "Which do I start with?"
    If you have a Jetson or a smaller GPU → start with **Cosmos-Reason2** (notebooks 01–05).
    If you have a big GPU and want video generation → jump to **Cosmos 3** (notebooks 06–07).

````python
# 🔧 Preflight — this cell never crashes. It just tells us what's available.
import os, sys, time
os.environ.setdefault("QWEN_VL_VIDEO_READER", "torchcodec")
sys.path.insert(0, os.path.abspath(".."))

# 🔒 Cosmos tools confine file access to a workspace allow-list for safety.
# The notebooks live in notebooks/ but our sample media sits one level up, so we
# explicitly widen the workspace to the project root + /tmp. (This is how you grant
# the tools access to a folder — never wider than you need.)
os.environ["COSMOS_WORKSPACE"] = os.pathsep.join([os.path.abspath(".."), "/tmp"])            # import strands_cosmos from repo root

PROJECT_ROOT = os.path.abspath("..")
SAMPLE_VIDEO = os.path.join(PROJECT_ROOT, "sample.mp4")
SAMPLE_IMAGE = os.path.join(PROJECT_ROOT, "sample.png")

def gpu_ready() -> bool:
    try:
        import torch
        return torch.cuda.is_available()
    except Exception:
        return False

READY = gpu_ready()
print("🎮 GPU available :", READY)
if READY:
    import torch
    print("   Device       :", torch.cuda.get_device_name(0))
    print(f"   VRAM         : {torch.cuda.get_device_properties(0).total_memory/1e9:.1f} GB")
else:
    print("   (CPU-only is fine for reading along — the heavy cells will skip safely.)")
print("📹 sample.mp4   :", os.path.exists(SAMPLE_VIDEO))
print("🖼  sample.png   :", os.path.exists(SAMPLE_IMAGE))
````
